```python
!pip install boto3
import json
from boto3.dynamodb.types import TypeDeserializer


deserializer = TypeDeserializer()
```

```python
def deserialize_dynamodb_item(item):
    """Converte um item do tipo DynamoDB JSON para um dicionário Python."""
    if isinstance(item, dict):
        return {k: deserialize_dynamodb_value(v) for k, v in item.items()}
    return item

def deserialize_dynamodb_value(value):
    """Converte um valor individual do tipo DynamoDB JSON para valor Python."""
    return deserializer.deserialize(value)

def parse_milestone_column(row_value):
    try:
        # Transforma a string em lista de objetos
        milestone_list = json.loads(row_value)
        # Converte cada item da lista usando o deserializer
        return [deserialize_dynamodb_value(item) for item in milestone_list]
    except Exception as e:
        print(f"Erro ao processar valor: {row_value}, erro: {e}")
        return None


```

```python
dados['MILESTONE_PARSED'] = dados['MILESTONE'].apply(parse_milestone_column)
```
